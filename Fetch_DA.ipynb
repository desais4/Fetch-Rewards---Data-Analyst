### 1. RECEIPTS

```python
# import receipts file
r = open("receipts.json", "r")

df_r = pd.DataFrame() #convert it to a dataframe

while True:
    try:           # read line by line
        line = r.readline()
        data = json.loads(line)
    
        temp = pd.json_normalize(data)
        df_r = df_r.append(temp)
    
    except:        # break if last line
        break 
```

```python
# rename columns
df_r.rename(columns={
                          '_id.$oid': '_id',
                          'createDate.$date':'createDate',
                          'dateScanned.$date': 'dateScanned',
                          'finishedDate.$date': 'finishedDate',
                          'modifyDate.$date': 'modifyDate',
                          'pointsAwardedDate.$date': 'pointsAwardedDate',
                          'purchaseDate.$date': 'purchaseDate'
                          }, 
                 inplace=True)

# reorder columns
column_names = ['_id', 'bonusPointsEarned', 'bonusPointsEarnedReason',
                'createDate', 'dateScanned', 'finishedDate', 'modifyDate',
                'pointsAwardedDate', 'pointsEarned', 'purchaseDate', 
                'purchasedItemCount', 'rewardsReceiptItemList', 
                'rewardsReceiptStatus', 'totalSpent', 'userId']

r = df_r.reindex(columns=column_names)

#unix to timestamp conversion
r['createDate'] = pd.to_datetime(r['createDate'],unit='ms')
r['dateScanned'] = pd.to_datetime(r['dateScanned'],unit='ms')
r['finishedDate'] = pd.to_datetime(r['finishedDate'],unit='ms')
r['modifyDate'] = pd.to_datetime(r['modifyDate'],unit='ms')
r['pointsAwardedDate'] = pd.to_datetime(r['pointsAwardedDate'],unit='ms')
r['purchaseDate'] = pd.to_datetime(r['purchaseDate'],unit='ms')

#copy original data
r_original = r.copy()

# remove rewardsReceiptItemList
r.drop('rewardsReceiptItemList', axis=1, inplace=True)
r
```

### 2. RECEIPTS_ITEM
#### (dividing receipts table)

```python
# isolate rewardsReceiptItemList field from the receipts table
ri = r_original[['_id', 'rewardsReceiptItemList']]

# change table from receipt-based to item-based
ri = ri.explode('rewardsReceiptItemList')
ri = ri.reset_index(drop=True)
ri
```

```python
# flatten rewardsReceiptItemList
ri = pd.concat(
    [ri, json_normalize(ri['rewardsReceiptItemList'])], axis=1)

ri.drop('rewardsReceiptItemList', axis=1,  inplace=True)
ri
```

### 3. USERS

```python
#import users file
u = open("users.json", "r")

df_u = pd.DataFrame() #convert it to a dataframe

while True:
    try:           # read line by line
        line = u.readline()
        data = json.loads(line)
    
        temp = pd.json_normalize(data)
        df_u = df_u.append(temp)
    
    except:        # break if last line
        break
```

```python
# rename columns
df_u.rename(columns={
                          '_id.$oid': '_id',
                          'createdDate.$date':'createdDate',
                          'lastLogin.$date': 'lastLogin',
                          }, 
                 inplace=True)

# reorder columns
column_names = ['_id', 'state', 'createdDate',
                'lastLogin', 'role', 'active', 'signUpSource']

u = df_u.reindex(columns=column_names)

#unix to timestamp conversion
u['createdDate'] = pd.to_datetime(u['createdDate'],unit='ms')
u['lastLogin'] = pd.to_datetime(u['lastLogin'],unit='ms')
u
```
